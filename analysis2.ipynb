```python
#pip install pandasql
import pandasql as pds
```

```python
t=pds.sqldf("""select * from tit;""",globals())
```

```python
type(t)
```

```python
pds.sqldf("""select * from tit limit 10;""",globals())#select top 10 *....
```

```python
pds.sqldf("""select * from tit order by age ;""",globals())#select top 10 *....
```

```python
pds.sqldf("""select * from tit order by age desc ;""",globals())#select top 10 *....
```

```python
pds.sqldf("""select * from tit order by age desc limit 7  ;""",globals())#select top 10 *....
```

```python
tit.sort_values(by='age',ascending=False)
```

```python
tit.sort_values(by='age',ascending=False).head(7)
```

```python
#show all the passengers who were in first class (use pandasql for first option,use filter on df )
pds.sqldf("""select * from tit where class='First' ;""",globals())
```

```python
tit[tit['class']=='First']
```

```python
plt.figure(figsize=(16,10))
x=sns.barplot(data=tit,x='pclass',y='survived',estimator='sum')
for con in x.containers:
    x.bar_label(con,fmt="%.1f")
plt.show()
```
